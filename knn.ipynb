````python
import pandas as pd
import numpy as np
import random
````

````python
# set random seed for reproducibility
np.random.seed(42)

#generate data
num_rows = 100

# creating randomd data
data = {
    'User ID': range(1, num_rows + 1),
    'Gender': [np.random.choice(['Male', 'Female']) for _ in range(num_rows)],
    'Age': [random.randint(18, 65) for _ in range(num_rows)],
    'Estimated Salary': [np.random.randint(20000, 200000) for _ in range(num_rows)],
    'Purchased': np.random.randint(0, 2, num_rows)
}

# laod data to dataframe
df = pd.DataFrame(data)
df.to_csv('knndataset.csv', index= False) # saving as a csv file
````

````python
dataset = pd.read_csv('knndataset.csv')

X = dataset.iloc[:, [2, 3]].values # features
y = dataset.iloc[:, 4].values # values

# display(X)
dataset.describe()
dataset['Age'].mode()
````

````python
# Splitting the dataset for training and testing
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
````

````python
# feature scaling
from sklearn.preprocessing import StandardScaler
st_x = StandardScaler()
X_train = st_x.fit_transform(X_train)
X_test = st_x.transform(X_test)
````

Fitting K-NN classifier to the training data: Now we will fit the K-NN classifier to the training data. To do this, we will import the `KNeighborsClassifier` class from sklearn's neighbors library. After importing the class, we will create the classifier object of the class. The parameters of this class will be:

```
n_neighbors: Defines the required number of neighbors for the algorithm. By default, it is set to 5.  
metric = 'minkowski': This is the default parameter, and it determines the distance metric to be used between the points.  
p = 2: It is equivalent to the standard Euclidean metric.
```

````python
# Fitting KNNclassifier to the training set
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
````

**Predicting**: To predict the test set results, we will create a `y_pred` vector, as we did in Logistic Regression.

````python
y_pred = classifier.predict(X_test)
y_pred
````
