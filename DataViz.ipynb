```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib.colors import LinearSegmentedColormap
```

1. Matplot  basic visualization

```python
# Line plot
x = np.linspace(0, 10, 100)
y =np.sin(x)
```

```python
plt.plot(x, y, label = "Sinusoidal wave")
plt.title("Sin Wave Example")
plt.legend()
plt.show()
```

```python
# Bar chart
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.bar(categories, values, color='pink')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
```

```python
#Histogram and KDE plot
data = np.random.randn(1000)
sns.histplot(data, kde=True, color = "pink")
plt.title("Histogram with KDE")
plt.show()
```

```python
# heatmap
pink_cmap = LinearSegmentedColormap.from_list("custom_pink", ["#ffe6f0", "#ff80bf", "#ff3385"])
matrix = np.random.rand(5, 5)
sns.heatmap(matrix, annot=True, cmap=pink_cmap, fmt=".2f")
plt.title("Heatmap Example")
plt.show()
```

```python
# pairplot
pink_palette = ["#ff99cc", "#ff66b3", "#ff3385"]
sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species", palette=pink_palette)
plt.show()
```
